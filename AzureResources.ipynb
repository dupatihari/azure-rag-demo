```python
!APP_NAME="mcp-campaign-insights" ; \
RG="rg-rag-demo" ; \
ENV_NAME="cae-rag-demo" ; \
ACR_NAME="acrragdemo19865" ; \
ACR_SERVER="acrragdemo19865.azurecr.io" ; \
IMAGE="$ACR_SERVER/mcp-campaign-insights:1" ; \
FUNCTION_URL="https://func-campaign-insights-uks01.azurewebsites.net/api/getcampaigninsights" ; \
ACR_USER=$(az acr credential show -n $ACR_NAME --query username -o tsv) ; \
ACR_PASS=$(az acr credential show -n $ACR_NAME --query "passwords[0].value" -o tsv) ; \
(az containerapp show -g $RG -n $APP_NAME >/dev/null 2>&1 && \
  az containerapp update -g $RG -n $APP_NAME \
    --image "$IMAGE" \
    --set-env-vars CAMPAIGN_INSIGHTS_FUNCTION_URL="$FUNCTION_URL" \
 || \
  az containerapp create -g $RG -n $APP_NAME \
    --environment $ENV_NAME \
    --image "$IMAGE" \
    --ingress external \
    --target-port 8000 \
    --registry-server "$ACR_SERVER" \
    --registry-username "$ACR_USER" \
    --registry-password "$ACR_PASS" \
    --env-vars CAMPAIGN_INSIGHTS_FUNCTION_URL="$FUNCTION_URL" \
    --min-replicas 1 --max-replicas 3 )
```
